1. Na lista anterior, colocamos que uma das formas de se aproximar o pi é através da soma
$$
\pi = 4 (\sum_n^\infty \frac {(-1)^n} {2n+1}).
$$
Dado um valor de $n$ qualquer, faça uma lista que armazene o valor dessa soma para cada valor de 0 até $n$ nos índices correspondentes (o índice 0 armazena o valor da soma para $n=0$, o índice 1 armazena o valor da soma para $n=1$,...). Ao final, use essa lista para plotar um gráfico com os valores da soma.

```python
#Possível resolução para a última
import numpy as np
from matplotlib import pyplot as plt

n = 100
sum = 0 
y=[]

for i in range(n):
  sum+= (-1)**i/(2*i+1)
  y.append(4*sum)


plt.plot(y)
```
