1. Uma das formas de se aproximar o número $\pi$ é utilizando a Fórmula de Leibniz, que nos dá que:
$$
1-\frac 1 3 + \frac 1 5 - \frac 1 7 + \frac 1 9 - ... = \frac \pi 4
$$
ou usando a notação de séries
$$
\sum_n^\infty \frac {(-1)^n} {2n+1} = \frac \pi 4.
$$
Faça um programa que calcule essa soma até um n-ésimo termo e retorne a aproximação de $\pi$ (lembrando de multiplicar por 4 ao final da série aproximada);
1. Um método de aproximar a raíz quadrada de um número $a$ qualquer é o Método de Newton, que é feito da seguinte forma: inicia-se com uma estimativa $y_0$, que pode ser inicialmente $\frac a 2$ por exemplo, e fazemos
$$
  y_1 = \frac {y_0 + \frac a {y_0}}{2} \\
  y_2 = \frac {y_1 + \frac a {y_1}}{2} \\
  \vdots \\
  y_n = \frac{ y_{n-1} + \frac {a} {y_{n-1}}}{2}
$$
onde $y_n$ é a estimativa para a raíz quadrada de $a$ após n iterações.
  1. Faça um programa que use o método acima para dar uma aproximação para a raíz quadrada de um número `a` qualquer informado com um número de `n` iterações, dado um `n` também informado, e apresente-a na tela no final.
  1. Faça um programa que use o método acima para dar uma aproximação para a raíz quadrada de um número `a` qualquer informado, até que que a diferença entre o a estimativa $y_n$ e a $y_{n-1}$ seja inferior a um `epsilon` informado, isto é, até que $|y_n - y_{n-1}| < epsilon$, e apresente-a na tela no final. Você pode usar a função `abs()` para obter o valor o absoluto (ou módulo) de um número.
