## Tipos de Dados

### Textos: 
Para textos temos o tipo de dado **String**. String são
cadeias de caracteres ordenados. Para definir uma String, como já visto anteriormente, colocamos o nosso texto entre aspas, que podem ser simples ( ‘ ) ou duplas ( “ ).
Exemplos: 
- `“Isso é uma String”`
- `'Isso também é uma String'`



### Números

Para números temos três tipos de dados:

#### Int
Int, como o nome sugere, são números inteiros. Podem ser negativos, positivos ou zero, e não possuem casas decimais. No Python não há um limite de tamanho para os int. Exemplo: - 
- `-12141241 é um Int. `
- `3 `
- `0 `
- `12141241`


#### Float

Floats são números negativos ou positivos que possuem uma ou mais casas decimais. Podemos representá-los com sua parte inteira, um ponto e sua parte decimal (Atenção: é necessário usar ponto ‘.’  para números, a vírgula possui outra função no Python). Exemplo: 152.1312. Também podemos colocar os números em notação científica (produto com potências de 10) especificando com um "`e`" e a potência desejada. Exemplo: `12.5e2` é equivalente a $12,5\times10^2 = 1250$, e `3.5e-2` é equivalente a $3,5*10^{-2} = 0.035$. Exemplos de float:
- `0.0`
- `-235.1`
- `0.5`
- `6434.5242`
- `1.3e4`





#### Complex

Complex são, como sugerido pelo nome, números complexos, ou seja, possuem uma parte real e uma parte imaginária. Representamos eles com a parte real, sinal de adição (+) (ou subtração (-) se a parte imaginária for negativa) e o valor da parte imaginária seguida de um `j`, que é a representação no Python da unidade imaginária ($\sqrt{-1}$). Mesmo quando a parte imaginária é 1 é preciso explicitá-la junto do `j`. Exemplos: 
- `4+3j.`
- `1j`
- `-2.3 - 5.4j`

Temos os valores possíveis `True`, que será o valor Verdadeiro, e o valor `False`, que é o valor Falso (note que a letra maiúscula do início é importante. Usaremos o Booleano ao tratar de expressões lógicas.


## Operadores

Os operadores, assim como na matemática, recebem um ou dois valores e retornam um terceiro de acordo com qual a operação está sendo efetuada. Temos os Operadores Matemáticos, que já estamos acostumados, os Operadores Lógicos, que usamos para operar valores lógicos (já vamos falar sobre), Operadores de Comparação, que são as relações de igualdade e desigualdade, e o Operador de Inclusão, usado quando lidando com tipos sequenciados.

### Operadores Matemáticos

Operador | Nome | Exemplo | Resultado
:---: | :---: | :---: | :---: 
+ | Adição | 2 + 3 | 5
-|Subtração|5.2 - 3.5|1.7
\* | Multiplicação |2.5 * 2 |5
/ |Divisão (racional) |11 / 2 |6.5
% |Módulo (resto da divisão de inteiros) |12%5 |2
\** |Exponenciação |2**3 |8
// |Divisão (inteira) |11 // 2 |5





### Operadores de Comparação

Operador | Nome | Exemplo | Resultado
:---: | :---: | :---: | :---: 
== | É igual a |5+2 == 7 |True 
!= | É diferente de | 3 != 1 + 2 |False
\> | É maior que | 3 > 2 | True
< | É menor que | 3 < 3 | False 
\>= | É maior ou igual | 3>=3 | True
<= | É menor ou igual | 2<=5 | True 



### Operadores Lógicos
Operador | Nome | Descrição | Exemplo | Resultado
:---: | :---: | :---: | :---: | :---: 
and | “e“ | Retorna verdadeiro se os dois valores forem verdade | 3>5 and 2 < 3 | False
or | “ou“ | Retorna verdadeiro se pelo menos um dos valores forem verdade | 4==2+2 or 2 \> 4 | True
not | “não“ | Retorna o valor inverso (False se for verdadeiro, True se for falso) | not 3 < 2 | True



### Operador de Inclusão
Operador | Descrição | Exemplo | Resultado
:---: | :---: | :---: | :---: 
in | Retorna verdadeiro se o item especificado está presente no objeto | 2 in [2,3,4] | True
not in | Retorna verdadeiro se o item especificado não está no objeto | 1 not in [2,3,4] | True





### Ordem dos operadores

Os operadores são executados na ordem apresentada abaixo. Operadores com mesma prioridade são lidos da esquerda pra direita.

1. Parênteses (dos mais internos pros mais externos)
1. Exponenciação
2. Multiplicação e divisão
3. Adição e subtração
3. Operadores de comparação
1. `not`
2. `and`
3. `or`


**Exemplo:** Vamos transcrever para a sintaxe do python a expressão numérica
$$
    5^2 \cdot \left[ \frac{1+1}{2\cdot 10} + \left( - \frac 1 4 \right) \right]
$$

```python
5**2 * ((1+1)/(2*10) + (-1/4)) 
```

### Operador de Concatenação (Strings)
Temos ainda com Strings o operador de concatenação, o “`+`”, que faz a junção de duas Strings em uma só. Exemplo:
- `"Olá, meu nome é: " + "Felipe"` tem como resultado a String `"Olá, meu nome é Felipe"`

É bom destacar que se não há espaço no final da primeira String ou no início da segunda, a String final também não terá. Então, `"Olá, meu nome é"+"Felipe"` tem como retorno a String `"Olá, meu nome éFelipe"`.

Importante reiterar também que o operador de concatenação só funciona com duas Strings. Se fizermos por exemplo:

- `"O valor de x é: " + 3`

o sistema retornará um erro, pois estamos tentando concatenar uma String com um Int. Para isso, temos que usar a função `str()` no valor que queremos concatenar, pois essa função faz a conversão de um valor de qualquer tipo para um int. O correto então ficaria:
- `"O valor de x é: " + str(3)`

cujo retorno é a string 

`"O valor de x é 3"`.

## Variáveis


Variáveis são espaços em que podemos armazenar valores de qualquer tipo. Para declarar uma variável no Python é necessário apenas escrever seu nome, seguido de um sinal de igual e o valor que ela conterá.
- ` contador = 1 `
- ` numeroEPar = True `
- ` nome = "João Silva" `

O Python possui tipagem dinâmica, o que significa que não é necessário especificar o tipo da variável. Fica a cargo do programador cuidar para não haver erro com conflito de tipos. É possível colocar valores de tipos diferentes em uma mesma variável, porém não é recomendado que seja feito, para evitar erros. 

Para usar o valor armazenado na variável basta apenas escrever o nome dela onde necessário. Exemplo:


```python
a = 5
b = 3
print(a + b) #retorno: 8
```

Podemos alterar o valor da variável sempre que quisermos, é necessário apenas atribuir um novo valor com o sinal de igual, que irá sobrescrever o valor antigo da variável. Exemplo:

```python
a = 2
print(a) #retorno: 2
a = 1
print(a) #retorno: 1
```

Ainda  alterar o valor da variável usando o próprio valor dela. Se quisermos por exemplo, somar 1 ao valor presente na variável, podemos escrever:
`a = a + 1`, o que armazena na variável a o valor presente nela antes somado a 1, o que pode ser feito pra qualquer outra operação com qualquer valor. Exemplo:


```python
a = 1
print("Valor inicial de a:")
print(a)
a = a + 1
print("Valor alterado de a:")
print(a)
```

Muitas vezes pode aparecer também a versão abreviada desse comando, escrito da seguinte forma, com o operador antes do sinal de igualdade:

` a += 1 `

Esse comando, que pode ser lido como "Some 1 ao valor presente na variável `a`" é exatamente o mesmo que `a = a + 1`, escrito apenas de forma abreviada.

Aqui estamos atribuindo valores especificados no código para as variáveis, mas é possível também pedir ao usuário que insira um valor durante a execução do programa e atribuí-la a uma variável. Para isso usamos a função `input`. Na função `input` colocamos dentro dos parênteses uma mensagem que será exibida na tela (similar ao print()), com a diferença que o programa será interrompido até que o usuário insira algum texto e pressione Enter. O texto que o usuário inserir será atribuído na variável especificada, como por exemplo:


```python
	nome = input("Insira seu nome: ")
```

Temos agora que o texto inserido pelo usuário no input está armazenado na variável `nome`. Podemos utilizá-lo em qualquer momento, como para um print por exemplo. Execute o código abaixo que imprimirá a mensagem "Seu nome é: " seguida do nome que você inseriu no comando acima. Veja que estamos usando a variável `nome` para imprimir esse valor.

```python
print("Seu nome é: " + nome)
```

**Importante**: sem as aspas, ao escrever `print(nome)` estamos apresentando o valor presente na variável `nome`. Para passar um texto específico é necessário colocá-lo entre aspas. Tente executar o seguinte código:


```python
nome = "Maria"
print(nome) # apresenta “Maria” na tela
```

E agora tente executar o seguinte código:

```python
	print("nome") # apresenta “nome” na tela
```

**Importante 2**: por padrão, o input armazena o valor na variável como uma String. Ou seja, se o usuário inserir um número (123 por exemplo), o que será armazenado é a String “123”, e não o Int 123, o que irá ocasionar erros ao fazer operações. Para isso, podemos realizar a conversão de tipos com as seguintes funções:

- `int(texto)` -> converte a String para um int
	-	Ex: `int("321")` retorna o int `321`; 
- `float(texto)` -> converte a String para um float
	-	Ex: `float("123.4") retorna o float `123.4`
- `complex(texto)` -> converte a String para um complex
  - Ex: `complex("5+3j")` retorna o complex `5+3j`

Como visto anteriormente, também é possível realizar o processo inverso, isto é, tendo um número podemos convertê-lo para uma String com a função `str(numero)`.
