# Numpy

O Numpy é um módulo que nos permite realizar operações mais avançadas mas de maneira simples com um tipo diferente de listas próprias da biblioteca: o NumPy Array. Esse módulo é interessante para trabalharmos com vetores e matrizes de forma matemática, contendo um grande leque de ferramentas da álgebra linear e geometria analítica, o que nos permite trabalhar com esses conceitos de forma simples e direta.

Ele tem muitas similaridades e funções análogas com o MATLAB. Há inclusive uma página na documentação oficial comparando as funções do módulo com as presentes no MATLAB, no artigo [NumPy for MATLAB users

## Importando

````python
import numpy as np
````

## O Numpy Array

A principal ferramenta do Numpy é uma estrutura de dados denominada Numpy Array. Ele funciona muito similar às listas que vimos anteriormente, mas possui algumas funções interessantes para se tratar listas como vetores ou matrizes. Para declarar um Numpy Array usamos a função `np.array(arr)`, onde passamos o parâmetro `arr` que pode ser uma lista ou tupla que queremos interpretar como um vetor ou matriz.

Exemplo:

````python
a = np.array([1,2,3])
A = np.array([[1,2],[3,4]])
print("Vetor a: ")
print(a)
print("Matriz A: ")
print(A)
````

Aqui entra o interessante de se trabalhar com o Numpy Array: podemos fazer operações entre arrays como se fosse um número qualquer, como se estivéssemos trabalhando com vetores em $\mathbb R ^n$, matrizes em $\mathbb R ^ {m\times n}$ ou ainda matrizes multi-dimensionais, com 3, 4,..., n dimensões.

Exemplo: Trabalhando em $\mathbb R^n$ (vetores de tamanho n com uma dimensão), façamos um vetor $a = (1,2,3)$ e $b = (2,2,2)$.

````python
a = np.array([1,2,3])
b = np.array([2,2,2])

def f(x): #Criando uma função que recebe um número e retorna 2x+3
    return 2*x+3

print("Vetor a:",a)
print("Vetor b:",b)
print("Multiplicação por escalar -3*a: ",-3*a)
print("Função f aplicada no vetor a: ",f(a))
print("Soma de vetores a+b:",a+b)
print("Multiplicação ponto-a-ponto: ", a*b)
print("Produto escalar: ", a@b)
````

Exemplo 2: Trabalhando em $M_{mxn}$ (Matrizes 2-D $mxn$), vamos fazer duas matrizes:

$$
A = \begin{bmatrix}
    1 && 2 \\
    3 && 4
\end{bmatrix}
,
B = \begin{bmatrix}
    2 && 2 \\
    2 && 2
\end{bmatrix}
$$

````python
A = np.array([[1,2],[3,4]])
B = np.array([[2,2],[2,2]])

y = np.array([1,2])

def f(x): #Criando uma função que recebe um número e retorna 2x+3
    return 2*x+3

print("Matriz a:")
print(A)

print("Matriz b:")
print(B)

print("Multiplicação por escalar -3*A: ")
print(-3*A)

print("Função f aplicada no vetor a: ")
print(f(A))

print("Soma de vetores a+b:")
print(A+B)

print("Multiplicação ponto-a-ponto: ")
print(A*B)

print("Produto matricial A.B :")
print(A@B)

print("Matriz vetor Ax:")
print(A@y)

print("Resolvendo o sistema Ax = b")
print(np.linalg.solve(A,y))
````

## Algumas funções Interessantes

Há centenas de funções interessantes do Numpy, a lista completa pode ser vista na [documentação oficial] (https://numpy.org/devdocs/reference/). Listamos algumas interessantes  e mais frequentes abaixo:

Sendo `A` e `b` numpy arrays:

- `np.arange(n)`: cria um array nx1 indo de 0 a n-1. Também pode ser incluido um início e um passo, igual a função `range(n)`. Nesse caso ainda aceita valores reais;
- `np.linspace(a,b,n)`: cria um array nx1 indo de `a` até `b` com $n$ elementos igualmente espaçados;
- `np.zeros((m,n))`: retorna um vetor mxn preenchido por zeros
- `np.ones((m,n))`: retorna um vetor mxn preenchido por 1
- `np.identity(n)` : retorna a matriz identidade $nxn$.
- `A.T`: retorna a transposta de A
- `np.shape(A)`: retorna a dimensão de A na forma de uma tupla

````python
A = np.array([[1,2],[0,0]])
b = np.array([1,2])

print('arange(10)')
print(np.arange(10), "\n")

print("linspace(0,2,20)")
print(np.linspace(0,2,10),"\n")

print("np.zeros(2,2)")
print(np.zeros((2,2)),"\n")

print("np.identity(3)")
print(np.identity(3),"\n")

print("A.T")
print(A.T,"\n")

print("shape(A)")
print(np.shape(A),"\n")
````

## Matplotlib

O Matplotlib é uma biblioteca de ferramentas para criação de gráficos, estáticos, animados ou até interativos, para visualização no Python. Com esse conjunto de ferramentas é possível criar visualizações gráficas dos mais variados tipos, desde a criação de gráficos cartesianos tradicionais, estatísticos como gráficos de barras ou histogramas, até exemplos mais avançados como gráficos tridimensionais e trabalhos com imagens. É uma das ferramentas mais básicas e intuitivas de se trabalhar com gráficos no Python, mas possui um acervo bem grande de coisas possíveis de se fazer e de personalizações que as ferramentas permitem.

Essa biblioteca é um módulo que naturalmente seria necessário instalar, através do *pip* ou outro gerenciador de pacotes, mas no Google Colab ele já vem instalado.

## Importando
Como comentado, o Matplotlib é uma biblioteca extensa e com muitos conjuntos de ferramentas diferentes. Assim, não compensa tanto importar todas as ferramentas, pois aloca uma certa quantidade de memória desnecessária no nosso programa, podendo apenas importar alguns dos conjuntos de ferramentas presentes no módulos usando o método `from - import` do Python.

Aqui iremos apenas importar o `pyplot`, conjunto de ferramentas do Matplotlib com o básico para se trabalhar com gráficos bidimensionais. Para importar o pyplot, podemos usar o comando abaixo:

````python
from matplotlib import pyplot as plt ## plt é um apelido comum para o pyplot, comum em muitos exemplos e programas
````

Os gráficos mais básicos do Matplotlib funcionam com base em dados presentes em Listas ou Dicionários. Podemos começar pelo básico por exemplo de gráficos cartesianos:

## Gráficos Cartesianos

O gráfico cartesiano no Matplotlib funciona no sistema de correspondência de coordenadas. Nós podemos passar como argumento para o método `plt.plot()` uma lista de valores `[y1,y2,...,yn]`, e o método assim irá plotar um gráfico de linhas conectando as coordenadas $(0,y_1),(1,y_2),...,(n-1,y_n)$ no plano (lembrando que a contagem de vetores no Python começa no 0, por isso com n valores as abcissas vão de 0 a n-1). Chamamos o método `plt.plot()` com a lista de valores como parâmetro, e em sequência chamamos o comando `plt.show()` para exibir o gráfico (tente alterar os valores de y para ver a alteração no gráfico):


````python
y = [3,4,1,2]

plt.plot(y)
plt.show()
````

Por padrão, passando apenas uma lista de n valores irá imprimir o gráfico com as abcissas de $0$ até $n-1$. Caso queiramos passar abcissas personalizadas, basta passar antes da lista de valores de `y` uma lista de valores `[x1,x2,...,xn]`. Dessa forma, as coordenadas do gráfico conectadas serão as coordenadas $(x_1,y_1),(x_2,y_2),...,(x_n,y_n)$.

**Importante:** é preciso atentar que as duas listas, `x=[x1,...,xn]` e `y=[y1,..,yn]`, tenham o mesmo tamanho $n$, e que as coordenadas correspondentes estejam nas posições correspondentes.

O código fica da seguinte forma:
```python
plt.plot([x1,...,xn],[y1,...,yn])
plt.show()
```
onde
- `[x1,...,xn]` é uma lista de valores para as abcissas (valores de x no gráfico);
- `[y1,...,yn]` é uma lista de valores para as ordenadas (valores de y no gráfico);
- `plt.plot()` é o método que recebe os valores a serem plotados;
- `plt.show()` é o método que exibe o gráfico com as configurações indicadas;

Nota: mesmo sem chamar o método `plt.show()` o gráfico pode ser exibido, porém, é sempre ideal executá-lo para encerrar as configurações do gráfico em questão e não confundir caso haja um gráfico na sequência.


````python
x = [0,1,2,3,4,5,6]

y1 = []
for i in x:
    y1.append(i**2)

y2 = []
for i in x:
    y2.append(2**i)

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
````

Por padrão o gráfico utilizado é o de linhas azuis, mas podemos personalizar passando um parâmetro a mais indicando o tipo de plot utilizado. Podemos colocar marcadores, linhas e indicar a cor usada. Por exemplo:
- `"-"` faz um gráfico unindo os pontos com uma linha contínua (é o padrão)
- `"."` faz um gráfico marcando com pontinhos
- `"o"` faz um gráfico com bolinhas
- `"--"` faz um gráfico com uma linha traçada
- `"-o"` faz um gráfico com uma linha contínua e bolinhas nos pontos plotados

A cor usada pode ser indicada na sequência final. Algumas cores são:
- 'b' azul
- 'g' verde
- 'r' vermelho
- 'c' ciano
- 'm' magenta
- 'y' amarelo
- 'k' preto
- 'w' branco

A documentação inteira pode ser vista na [seguinte página](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot).

````python
x = [0,1,2,3,4,5,6]
y1 = [i**2 for i in range(7)]

plt.plot(x,y1,".")
plt.show()


plt.plot(x,y1,"or")
plt.show()

````

````python
plt.plot(x,y1,"--g")
plt.show()

plt.plot(x,y1,"-oy")
plt.show()
````

## Aliando Numpy e Matplotlib

O Numpy e o Matplotlib são dois módulos que funcionam perfeitamente em conjunto, tendo funções em ambas que funcionam uma com a outra, podendo usar Numpy Arrays para a plotagem dos gráficos.

**Exemplo:** plotagem de funções. Façamos a plotagem da função $x^2-5x+2$ no intervalo $[-3,5]$.

Precisamos fazer um vetor com o intervalo em que se deseja plotar a função, e em seguida aplicar a função nesse vetor do intervalo.

Podemos usar a função `np.linspace(a,b,n)` do Numpy para fazer o intervalo desejado, tendo `a` o ponto de início, `b` o ponto final e `n` o número de elementos no vetor. No nosso caso temos `a=-3`, `b=5` e `n` é arbitrário, mas quanto menores valores de `n` mais "quadrado" vai ficar nosso gráfico, e para valores de `n` muito grandes pode pesar um pouco no processamento. Vale testar para um certo número de valores para ver a partir de que número de pontos que o gráfico suaviza.

````python
x = np.linspace(-3,5,20)
print(x)
````

Como visto, podemos fazer uma função que recebe um Numpy Array e trata como um número qualquer, que vai retornar um vetor com a função aplicada em todos os valores do array passado por parâmetro.

Com a função criada, podemos aplicar no array `x` do intervalo que criamos.

````python
def f(x):
    return x**2 - 5*x + 2

y = f(x)
print(y)
````

Agora podemos plotar a função usando o `pyplot`:

````python
plt.plot(x,y)
plt.show()
````

## Manipulando a planilha

O *pandas* trabalha com duas estruturas:

- As **Series**, que são sequências de dados que funcionam como as listas. Ao importar a planilha com o *pandas*, as colunas dessa planilha serão interpretadas como as Series;
- Os **DataFrames**, que agregam diferentes Series, que é basicamente nossa planilha inteira.

Para acessar o dataframe mexemos simplemente com a variável em que armazenamos a planilha. Para acessar uma Series, acessamos a variável da planilha indexada pelo nome das Series, que por padrão é o nome presente no cabeçalho de cada coluna.

Exemplo: podemos acessar a lista de nomes das cidades presentes na planilha do IDH. Selecionamos através do nome da coluna, que é indicada como `"Município"`.

````python
dadosidh["Município"]
````

E no caso de dados numéricos, podemos operar sobre os Series usando os operadores básicos do Python (matemáticos, de comparação, etc). O resultado disso será uma nova Series com a operação aplicada entrada por entrada correpondente (na mesma linha).

Exemplo: queremos criar uma taxa que é a razão entre a Renda per capita e a Expectativa de anos de estudo.

````python
dadosidh["Renda per capita"] / dadosidh["Expectativa de anos de estudo"]
````

Podemos ainda adicionar essa nova Series à planilha, criando um novo índice, como se fossemos acessá-lo, e atribuindo o resultado da expressão a esse novo índice. Seguindo o exemplo anterior, vamos criar uma coluna chamada "Renda por anos de estudo" com o resultado da expressão acima.

````python
dadosidh["Renda por anos de estudo"] = dadosidh["Renda per capita"] / dadosidh["Expectativa de anos de estudo"]
````

Vejamos agora a planilha atualizada, e perceba que agora nossa nova coluna:
